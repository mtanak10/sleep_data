I will also change the blood pressure column as it includes two values: systolic and diastolic pressure as systolic_BP and diastolic_BP, respectively. 

Since we are aiming to see if the person is overweight/obese or not, I will change the BMI category. Overweight and obese were saved as 1, and other results were changed to 0. I will name the that column as bmi_category_binary. 

```r
sleep_data <- sleep_data |> 
    drop_na() |>
    separate(blood_pressure, into = c("systolic_BP", "diastolic_BP"), sep = "/") |>
    mutate(
        systolic_BP = as.numeric(systolic_BP),
        diastolic_BP = as.numeric(diastolic_BP),
        bmi_category_binary = if_else(bmi_category %in% c("Overweight", "Obese"), 1, 0),
        bmi_category_binary = as.factor(bmi_category_binary)
    ) 
    
    
    
head(sleep_data)

colSums(is.na(sleep_data))
```

```r
sleep_plot <- ggplot(sleep_data, aes(x = physical_activity_level, y = heart_rate, color = bmi_category_binary)) +
    geom_point() +
    geom_jitter() + 
    labs(
        title = "Physical Activity Level vs Heart Rate Colored by the BMI category",
        x = "Physical Activity Level (minutes)", 
        y = "Heart Rate (bpm)",
        color = "BMI Category 1: Obese or Overweight, 0: Normal"
    ) +
    scale_x_continuous(expand = expansion(mult = c(0.1, 0.2))) +  
    theme_minimal()

sleep_plot
```
